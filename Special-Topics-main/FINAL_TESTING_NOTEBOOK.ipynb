```python
import nbformat

d3_path = PROJECT_ROOT / "d3_graphrag" / "Special Topics D3.ipynb"

nb = nbformat.read(d3_path, as_version=4)

for cell in nb.cells:
    if cell.cell_type == "code":
        code = cell.source
        
        # skip shell/magic cells
        if code.strip().startswith(("!", "%")):
            continue
        
        try:
            exec(code, globals())
        except Exception as e:
            print("Skipped one cell because of error:")
            print(type(e).__name__, e)
            continue

print("D3 notebook loaded as much as possible.")
```

## 3. Test hybrid retrieval

```python
test_question = "What is retrieval augmented generation?"

hybrid_results = hybrid_search(test_question, top_k=5)

print("Hybrid results:", len(hybrid_results))

for i, r in enumerate(hybrid_results, 1):
    print("\n" + "="*80)
    print(f"RESULT {i}")
    print("TITLE:", r.get("title", "N/A"))
    print("PAGE:", r.get("page", "N/A"))
    print("PAPER ID:", r.get("paper_id", "N/A"))
    print("TEXT:", str(r.get("text", ""))[:500])
```

## 4. Test GraphRAG retrieval

This is the main live-demo test. It should return graph-expanded chunks with page numbers.


```python
question = "What is retrieval augmented generation?"

graphrag_result = graphrag_retrieve(question, top_k=5)

print("Question:", graphrag_result["query"])
print("Expanded paper IDs:", graphrag_result["expanded_paper_ids"])
print("Graph context chunks:", len(graphrag_result["graph_context"]))

for i, c in enumerate(graphrag_result["graph_context"], 1):
    print("\n" + "="*80)
    print(f"GRAPH CONTEXT {i}")
    print("TITLE:", c.get("title", "N/A"))
    print("PAGE:", c.get("page", "N/A"))
    print("CHUNK ID:", c.get("chunk_id", "N/A"))
    print("TEXT:", str(c.get("text", ""))[:700])
```
